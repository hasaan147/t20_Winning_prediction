```python
from sklearn.preprocessing import LabelEncoder
le_f=LabelEncoder()
le_a=LabelEncoder()
le_ca=LabelEncoder()
le_e=LabelEncoder()
le_cy=LabelEncoder()
le_f=LabelEncoder()
le_car=LabelEncoder()
```

```python
df['fueltype']=le_f.fit_transform(df['fueltype'])
df['aspiration']=le_a.fit_transform(df['aspiration'])
df['carbody']=le_ca.fit_transform(df['carbody'])
df['enginetype']=le_e.fit_transform(df['enginetype'])
df['cylindernumber']=le_cy.fit_transform(df['cylindernumber'])
df['fuelsystem']=le_f.fit_transform(df['fuelsystem'])
df['CarName']=le_car.fit_transform(df['CarName'])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerical_features = ['symboling', 'enginesize', 'horsepower', 'citympg', 'highwaympg']

for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'{feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()
```

```python
categorical_features = ['fueltype', 'aspiration', 'carbody', 'enginetype', 'cylindernumber', 'fuelsystem']

for feature in categorical_features:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=df[feature])
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
numerical_features = ['symboling', 'enginesize', 'horsepower', 'citympg', 'highwaympg', 'price']

for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[feature])
    plt.title(f'Box plot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

```python
def remove_outliers(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
    return df_cleaned
```

```python
from sklearn.model_selection import train_test_split
x=df.drop("price",axis=1)
y=df.price
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
features_to_clean = ['enginesize', 'horsepower', 'highwaympg']

for feature in features_to_clean:
    df = remove_outliers(df, feature)
df
```

```python
for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[feature])
    plt.title(f'Box plot of {feature} (After Outlier Removal)')
    plt.xlabel(feature)
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
